```python
from mech_int.neel_digits import *
from mech_int import neel_digits
import seaborn as sns
sns.set()
plt.rcParams['figure.figsize'] = (15, 3)
```

```python
d_model = 64
d_mlp = 64
num_epochs = 50_000
LOSS = {}
MODELS = {}
```

```python
train(is_finite=False, d_model=d_model, d_mlp=d_mlp, num_epochs=num_epochs,
      seed=134526, plot=False, checkpoint_every=1000)
LOSS['original'] = neel_digits.train_losses
MODELS['original'] = neel_digits.model
```

```python
pd.DataFrame(np.stack(neel_digits.per_token_losses_list, axis=0)).plot()
```

```python
train(is_finite=False, d_model=d_model, d_mlp=d_mlp, num_epochs=num_epochs,
      mlp_bias=False,
      seed=134526, plot=False, checkpoint_every=1000)
LOSS['no_mlp_const'] = neel_digits.train_losses
MODELS['no_mlp_const'] = neel_digits.model
```

```python
train(is_finite=False, d_model=d_model, d_mlp=d_mlp, num_epochs=num_epochs,
      mlp_bias=False, output_proj=False,
      seed=134526, plot=False, checkpoint_every=1000)
LOSS['no_output_proj'] = neel_digits.train_losses
MODELS['no_output_proj'] = neel_digits.model
```

```python
train(is_finite=False, d_model=d_model, d_mlp=d_mlp, num_epochs=num_epochs,
      mlp_bias=False, output_proj=False, resid_mlp=False,
      seed=134526, plot=False, checkpoint_every=1000)
LOSS['no_resid_mlp'] = neel_digits.train_losses
MODELS['no_resid_mlp'] = neel_digits.model
```

```python
train(is_finite=False, d_model=d_model, d_mlp=d_mlp, num_epochs=num_epochs,
      mlp_bias=False, output_proj=False, resid_mlp=False, resid_attn=False,
      seed=134526, plot=False, checkpoint_every=1000)
LOSS['no_resid_attn'] = neel_digits.train_losses
MODELS['no_resid_attn'] = neel_digits.model
```

```python
pd.concat({label: pd.Series(loss) for label, loss in LOSS.items()}, axis='columns').plot(alpha=0.7)
```

```python
for label, model in MODELS.items():
    print(label, sum(t.numel() for t in model.state_dict().values()))
```
